```python
import pandas as pd
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
titanic = pd.read_csv('train (3).csv')
```

```python
titanic.head()
```

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
iris = sns.load_dataset('iris')
```

```python
#scatterplot(numerical-numerical)
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue=tips['sex'],style=tips['smoker'],data=tips)
```

```python
#barplot(numerical-categorical)
titanic.head()
```

```python
sns.barplot(x=titanic['Pclass'],y=titanic['Age'],hue=titanic['Sex'])
```

```python
#boxplot(numerical-categorical)
sns.boxplot(x=titanic['Sex'],y=titanic['Age'],hue=titanic['Survived'],data=titanic)
```

```python
#distplot(numerical-categorical)
sns.distplot(x=titanic[titanic['Survived']==0]['Age'],hist=False)
sns.distplot(x=titanic[titanic['Survived']==1]['Age'],hist=False)
```

```python
#heatmap(categorical-categorical)
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
titanic.groupby(['Pclass']).mean()['Survived']
```

```python
#clustermap(categorical-categorical)
sns.clustermap(pd.crosstab(titanic['SibSp'],titanic['Survived']))
```

```python
#pairplot (numerical-numerical graph for each value)
sns.pairplot(iris,hue='species')
```
